```python
import transformers
from transformers import BertModel, AutoTokenizer
import numpy as np
import pandas as pd
import torch
from datasets import load_dataset
from warnings import filterwarnings
from torch.utils.data import Dataset
import torch

filterwarnings("ignore")
```

```python
data = pd.read_csv("/kaggle/input/posts-final-project/posts.csv").drop("Unnamed: 0", axis=1)
data.head()
```

```python
from cleantext import clean

new_texts = []

text = data["text"]

for i in range(len(text)):
    new_text = clean(text[i], no_urls=True, no_emails=True,
                     no_phone_numbers=True, no_currency_symbols = True,
                     no_punct=True, no_emoji=True, no_line_breaks=True)
    
    new_texts.append(new_text)
```

```python
from transformers import BertTokenizer, BertModel

tokenizer = BertTokenizer.from_pretrained('bert-base-cased')
model = BertModel.from_pretrained("bert-base-cased")

encoded_input = tokenizer(new_texts, return_tensors='pt', truncation=True, padding = True)
```

```python
embs = encoded_input["input_ids"]
```
